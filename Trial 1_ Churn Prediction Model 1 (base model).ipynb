# Cleaning dataset

## Dropping unnecessary columns

```python
df.drop(columns=["CustomerID", "Churn Category", "Churn Reason"], inplace=True)
```

## Label Encoding

```python
print("Unique values of categorical columns\n")
for col in df.columns:
  if not df[col].dtype in ["int64", "float64"]: 
    print(f"{col}:")
    print(f"\tUniques: {df[col].unique()}")
    print(f"\tNumber of uniques: {df[col].nunique()}\n")
```

```python
# label encoding

from sklearn.preprocessing import LabelEncoder
cols = ["Gender", "Married", "Phone Service", "Internet Service", "Paperless Billing"]
le = LabelEncoder()
for col in cols:
  df[col] = le.fit_transform(df[col])
```

```python
df.head()
```

## Splitting dataset into X and y

```python
X = df.iloc[:, :-1]
y = df["Customer Status"]
```

```python
X
```

## Encoding dependent variable

```python
y
```

```python
# encoding dependent variable
# (Stayed = 0, Churned = 1)

y = y.apply(lambda x: 0 if x == "Stayed" else 1)
```

## One hot encoding

```python
# one hot encoding

X = pd.get_dummies(X)
```

## Splitting dataset into train, test and val

```python
# splitting into train (60%), test (20%) and val (20%)

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4)
X_test, X_val, y_test, y_val = train_test_split(X_test, y_test, test_size=0.5)
```

## Feature scaling

```python
# feature scaling

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
X_val = scaler.transform(X_val)
```

```python
X_train
```

```python
X_test
```

```python
X_val
```

# Training

## Building model

```python
# building the model

import tensorflow
from tensorflow.keras.layers import Dense, BatchNormalization
from tensorflow.keras.models import Sequential

model = Sequential(layers=[
    Dense(64, activation="relu", input_shape=(25,)),
    BatchNormalization(),
    Dense(128, activation="relu"),
    BatchNormalization(),
    Dense(128, activation="relu"),
    BatchNormalization(),
    Dense(20, activation="relu"),
    BatchNormalization(),
    Dense(1, activation="sigmoid")
])
```

## Compiling and saving

```python
model.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])
```

```python
checkpointer = tensorflow.keras.callbacks.ModelCheckpoint(
    "Trial 1: model1_weights.hdf5", 
    monitor="val_loss", 
    verbose=1, 
    save_best_only=True, 
    save_weights_only=True
)

earlystopping = tensorflow.keras.callbacks.EarlyStopping(
    monitor="val_loss",
    patience=20,
    verbose=1
)
```

```python
with open("Trial 1: model1.json", "w") as json_file:
  json_file.write(model.to_json())
```

## Fitting

```python
history = model.fit(
    X_train, 
    y_train, 
    batch_size=8, 
    validation_data=(X_val, y_val), 
    verbose=1,
    callbacks=[checkpointer, earlystopping],
    shuffle=True,
    epochs=100
)
```

# Analysing

```python
training_acc = history.history["accuracy"]
training_loss = history.history["loss"]
val_acc = history.history["val_accuracy"]
val_loss = history.history["val_loss"]
```

```python
plt.figure(figsize=(7, 7))
plt.plot(history.epoch, training_acc, label="Training Accuracy")
plt.plot(history.epoch, val_acc, label="Validation Accuracy")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(7, 7))
plt.plot(history.epoch, training_loss, label="Training Loss")
plt.plot(history.epoch, val_loss, label="Validation Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.show()
```

```python
y_pred = np.round(model.predict(X_test))
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report, precision_score, recall_score

cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, cbar=False)
plt.show()
```

```python
accuracy_score(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```

```python
stayed_precision = precision_score(y_test, y_pred, labels=[0, 1], pos_label=0)
```

```python
churned_precision = precision_score(y_test, y_pred, labels=[0, 1], pos_label=1)
```

```python
stayed_recall = recall_score(y_test, y_pred, labels=[0, 1], pos_label=0)
```

```python
churned_recall = recall_score(y_test, y_pred, labels=[0, 1], pos_label=1)
```

```python
plt.bar(["0", "1"], [stayed_precision, churned_precision])
plt.xlabel("Class")
plt.ylabel("Precision")
plt.title("Precision (0 = Stayed, 1 = Churned)")
plt.show()
```

```python
plt.bar(["0", "1"], [stayed_recall, churned_recall])
plt.xlabel("Class")
plt.ylabel("Recall")
plt.title("Recall (0 = Stayed, 1 = Churned)")
plt.show()
```

Precision and recall for Class 1 (Churned) is significantly lower than the precision and recall for Class 0 (Stayed) respectively.

This is most likely due to the imbalanced nature of the dataset. There are more observations belonging to the majority class (Stayed) and less belonging to the minority class (Churned).
